```python
data = pd.read_csv("../input/breast-cancer-wisconsin-data/data.csv")
```

```python
data=data.drop(labels=["Unnamed: 32"],axis=1)
```

```python
data.info()
```

```python
data["diagnosis"]=data["diagnosis"].map({"M":1,"B":0})
```

```python
sns.countplot(data["diagnosis"],label="Count")
```

```python
#Split data into mean, se and worst

features_mean=list(data.columns[2:12])
features_se=list(data.columns[12:22])
features_worst=list(data.columns[22:32])
print(features_mean)
print(features_se)
print(features_worst)
```

```python
data.columns
#data.shape
```

```python
corr=data[features_mean].corr()
plt.figure(figsize=(14,14))
sns.heatmap(corr, cbar = True,  square = True, annot=True, fmt= '.2f',annot_kws={'size': 15},
           xticklabels= features_mean, yticklabels= features_mean,
           cmap= 'coolwarm')
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(data[data.columns[1:32]],data["diagnosis"],test_size=0.3)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100)
model.fit(X_train,Y_train)
prediction=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(accuracy_score(Y_test, prediction))
```

```python
confusion_matrix(Y_test,prediction)
```

```python
print(classification_report(Y_test,prediction))
```

```python
from sklearn import svm
model = svm.SVC()
model.fit(X_train,Y_train)
prediction=model.predict(X_test)
accuracy_score(prediction,Y_test)
```

```python
print(confusion_matrix(Y_test,prediction))
print(classification_report(Y_test,prediction))
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='liblinear')
```

```python
classifier.fit(X_train,Y_train)
```

```python
prediction_2=classifier.predict(X_test)
print(accuracy_score(Y_test, prediction_2))
```

```python
confusion_matrix(prediction_2,Y_test)
print(classification_report(Y_test,prediction_2))
```

```python
param_grid={'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001],'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(svm.SVC(),param_grid,refit=True,verbose=4)
grid.fit(X_train,Y_train)
grid.best_params_
```

```python
grid_predict = grid.predict(X_test)
print(accuracy_score(Y_test, grid_predict))
```
